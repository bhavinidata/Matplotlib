```python
# Survival Rates
drug_g_mice_df = combined_df.groupby(["Drug", "Timepoint"])["Mouse ID"].count()
drug_g_mice_df = drug_g_mice_df.reset_index()
drug_g_mice_df.head()
```

```python
drug_g_mice_cap_df = drug_g_mice_df.loc[drug_g_mice_df["Drug"] == "Capomulin"]
drug_g_mice_cap_df["Mouse Survival Rate"] = (drug_g_mice_cap_df["Mouse ID"]/drug_g_mice_cap_df["Mouse ID"].max())*100
cap_survival_rate = drug_g_mice_cap_df["Mouse Survival Rate"].values
cap_survival_rate
```

```python
drug_g_mice_inf_df = drug_g_mice_df.loc[drug_g_mice_df["Drug"] == "Infubinol"]
drug_g_mice_inf_df["Mouse Survival Rate"] = (drug_g_mice_inf_df["Mouse ID"]/drug_g_mice_inf_df["Mouse ID"].max())*100
inf_survival_rate = drug_g_mice_inf_df["Mouse Survival Rate"].values
```

```python
drug_g_mice_ket_df = drug_g_mice_df.loc[drug_g_mice_df["Drug"] == "Ketapril"]
drug_g_mice_ket_df["Mouse Survival Rate"] = (drug_g_mice_ket_df["Mouse ID"]/drug_g_mice_ket_df["Mouse ID"].max())*100
ket_survival_rate = drug_g_mice_ket_df["Mouse Survival Rate"].values
```

```python
drug_g_mice_pla_df = drug_g_mice_df.loc[drug_g_mice_df["Drug"] == "Placebo"]
drug_g_mice_pla_df["Mouse Survival Rate"] = (drug_g_mice_pla_df["Mouse ID"]/drug_g_mice_pla_df["Mouse ID"].max())*100
pla_survival_rate = drug_g_mice_pla_df["Mouse Survival Rate"].values
```

```python
# fig, ax = plt.subplots()
# ax.set_xlim(-3, 48)
# ax.set_ylim(33,103)
# ax.set_title("Survival During Treatment")
# ax.set_xlabel("Time (Days)")
# ax.set_ylabel("Survival Rate (%)")
plt.xlim(-3, 48)
plt.ylim(33,103)
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
capomulin, = plt.plot(cap_timepoint, cap_survival_rate, "-o", color="red",alpha = 0.75, linewidth = 0.5, label = "Capomulin")
infubinol, = plt.plot(inf_timepoint, inf_survival_rate, "-^b",alpha = 0.75, linewidth = 0.5, label = "Infubinol")
ketapril, = plt.plot(ket_timepoint, ket_survival_rate, "-sg",alpha = 0.75, linewidth = 0.5, label = "Ketapril")
placebo, = plt.plot(pla_timepoint, pla_survival_rate,  "-dk",alpha = 0.75, linewidth = 0.5, label = "Placebo")
plt.legend(handles = (capomulin, infubinol, ketapril, placebo),  loc = "lower left")
plt.grid()
plt.show()
```

```python
drug_group_tumor_vol_mean_cap = drug_group_tumor_vol_mean.loc[drug_group_tumor_vol_mean["Drug"]=="Capomulin"]
drug_group_tumor_vol_mean_cap
initial_tumor_vol_cap = drug_group_tumor_vol_mean_cap.loc[0,"Tumor Volume (mm3)"]
initial_tumor_vol_cap
last_tumor_vol_cap = drug_group_tumor_vol_mean_cap.iloc[-1,2]
last_tumor_vol_cap
perc_change_tumor_vol_cap = ((last_tumor_vol_cap-initial_tumor_vol_cap)/initial_tumor_vol_cap)*100
perc_change_tumor_vol_cap
```

```python
drug_group_tumor_vol_mean_inf = drug_group_tumor_vol_mean.loc[drug_group_tumor_vol_mean["Drug"]=="Infubinol"]
drug_group_tumor_vol_mean_inf
initial_tumor_vol_inf = drug_group_tumor_vol_mean_inf.iloc[0,2]
initial_tumor_vol_inf
last_tumor_vol_inf = drug_group_tumor_vol_mean_inf.iloc[-1,2]
last_tumor_vol_inf
perc_change_tumor_vol_inf = ((last_tumor_vol_inf-initial_tumor_vol_inf)/initial_tumor_vol_inf)*100
perc_change_tumor_vol_inf
```

```python
drug_group_tumor_vol_mean_ket = drug_group_tumor_vol_mean.loc[drug_group_tumor_vol_mean["Drug"]=="Ketapril"]
drug_group_tumor_vol_mean_ket
initial_tumor_vol_ket = drug_group_tumor_vol_mean_inf.iloc[0,2]
initial_tumor_vol_ket
last_tumor_vol_ket = drug_group_tumor_vol_mean_ket.iloc[-1,2]
last_tumor_vol_ket
perc_change_tumor_vol_ket = ((last_tumor_vol_ket-initial_tumor_vol_ket)/initial_tumor_vol_ket)*100
perc_change_tumor_vol_ket
```

```python
drug_group_tumor_vol_mean_pla = drug_group_tumor_vol_mean.loc[drug_group_tumor_vol_mean["Drug"]=="Placebo"]
drug_group_tumor_vol_mean_pla
initial_tumor_vol_pla = drug_group_tumor_vol_mean_pla.iloc[0,2]
initial_tumor_vol_pla
last_tumor_vol_pla = drug_group_tumor_vol_mean_pla.iloc[-1,2]
last_tumor_vol_pla
perc_change_tumor_vol_pla = ((last_tumor_vol_pla-initial_tumor_vol_pla)/initial_tumor_vol_pla)*100
perc_change_tumor_vol_pla
```

```python
per_change = [perc_change_tumor_vol_cap, perc_change_tumor_vol_inf,
             perc_change_tumor_vol_ket, perc_change_tumor_vol_pla]
per_change
```

```python
x_axis = np.arange(len(per_change))
x_axis
```

```python
# help(plt.bar)
```

```python
plt.title("Tumor change Over 45 Day Treatment")
plt.ylabel("% Tumor volume change")
plt.ylim(-29, 69)
colors = ["g","r","r","r"]
rects=plt.bar(x_axis, per_change, color=colors, width = -1, alpha=0.90, align="edge")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Capomulin","Infubinol","Ketapril","Placebo"])
plt.axhline(0, color="k", alpha=0.5)
plt.grid()
def autolabel(rects, ax):
    width = -0.6
    for rect in rects:
        height = rect.get_height()
        if height<0:
            label_position = -8
        else:
            label_position = 7

        plt.text(x= width, y=label_position, s= "%d" %height.astype(int) + "%", fontsize = 14, color="white")
        width=width+1
autolabel(rects, per_change)
plt.tight_layout()
plt.savefig("../Images/avg_state_rain.png")
plt.show()
```
